```python
import cv2
```

```python
# Our Image
image_file = "car-image.jpg"
```

```python
# Our pre-trianed car classifier
classifier_file = "car-detector.xml"
```

```python
# create opencv image
img = cv2.imread(image_file)
img
```

```python
# convert to grayscale (need for haar cascade)
black_n_white = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
black_n_white
```

```python
# Display img in window
# display img
cv2.imshow("Shahid Car detector img", img)

# Dont autoclose (wait for key press)
cv2.waitKey(-1)
```

```python
# Display image in Jupyter
# Works on jupyter

from matplotlib import pyplot as plt

# Show color img
plt.imshow(img)
plt.title('Shahid Car Detector Color')
plt.show()

# Show grayscal img
plt.imshow(black_n_white)
plt.title('Shahid Car Detector Gray Scale')
plt.show()
```

```python
# create our classifier

car_tracker = cv2.CascadeClassifier(classifier_file)
car_tracker
```

```python
# Detect cars

# multiscale means detect car of any size or of any scale
#  This function will return a rectangle with coordinates(x,y,w,h) around the detected cars .
cars = car_tracker.detectMultiScale(black_n_white)
cars
```

```python
car1 =cars[0]
car1
```

```python
# Unpacking array

(x, y, w, h) = car1
print("x", x)
print("y", y)
print("w", w)
print("h", h)
```

```python
# Draw rectangle around a car
# cv2.rectangle(image, start_point, end_point, color, thickness)
cv2.rectangle(
    black_n_white,
    (x, y),
    (x + w, y + h),
    (255,0,0), # BGR
    2
)
```

```python
plt.imshow(black_n_white)
plt.title('Shahid Car Detector Gray Scale detected')
plt.show()
```

```python
# Draw rectangles around all cars
for(x, y, w, h) in cars:
    cv2.rectangle(
    black_n_white,
    (x, y),
    (x + w, y + h),
    (255,0,0), # BGR
    2
    )
    
plt.imshow(black_n_white)
```

```python
# Create a VideoCapture object and read from input file
# If the input is taken from the camera, pass 0 instead of the video file name.
video = cv2.VideoCapture('../Tesla Autopilot Dashcam Compilation 2021 Version-d4L1Pte7zVc.mp4')
video
```

```python
# Runs until car stops
while True:
    
    # read the current frame
    (read_successful, frame) = video.read()
    
    if read_successful:
        gray_scaled_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    else:
        break
    
    cars = car_tracker.detectMultiScale(gray_scaled_frame)
    # display img
    for(x, y, w, h) in cars:
        cv2.rectangle(
        frame,
        (x, y),
        (x + w, y + h),
        (255,0,0),
        2
        )
        
    cv2.imshow("Shahid Car detector img", frame)

    # Dont autoclose (wait for key press)
    cv2.waitKey(1)
```

```python
# Works here on Jupyter Notebook
from IPython.display import Video


# Runs until car stops
while(video.isOpened()):
    
    # read the current frame
    (read_successful, frame) = video.read()
    
    if read_successful:
        gray_scaled_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    else:
        break
    
    cars = car_tracker.detectMultiScale(gray_scaled_frame)
    # display img
    for(x, y, w, h) in cars:
        cv2.rectangle(
        frame,
        (x, y),
        (x + w, y + h),
        (255,0,0),
        2
        )
    
    cv2.imshow("Shahid Car detector img", frame)

    # Dont autoclose (wait for key press)
    if cv2.waitKey(25) & 0xFF == ord('q'):
      break
#     cv2.waitKey(1)
```

```python

cap = cv2.VideoCapture('../Tesla Autopilot Dashcam Compilation 2021 Version-d4L1Pte7zVc.mp4')
# Check if camera opened successfully
if (cap.isOpened()== False): 
  print("Error opening video stream or file")

# Read until video is completed
while(cap.isOpened()):
  # Capture frame-by-frame
  ret, frame = cap.read()
  if ret == True:

    # Display the resulting frame
    cv2.imshow('Frame',frame)

    # Press Q on keyboard to  exit
    if cv2.waitKey(25) & 0xFF == ord('q'):
      break

  # Break the loop
  else: 
    break

# When everything done, release the video capture object
cap.release()

# Closes all the frames
cv2.destroyAllWindows()
```
