```python
import cv2
```

```python
# Our Image
image_file = "car-image.jpg"
```

```python
# Our pre-trianed car classifier
classifier_file = "car-detector.xml"
```

```python
# create opencv image
img = cv2.imread(image_file)
img
```

```python
# convert to grayscale (need for haar cascade)
black_n_white = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
black_n_white
```

```python
# it works on terminal not on jupyter

# display img
cv2.imshow("Shahid Car detector img", img)

# Dont autoclose (wait for key press)
cv2.waitKey()
```

```python
# Works on jupyter

from matplotlib import pyplot as plt

# Show color img
plt.imshow(img)
plt.title('Shahid Car Detector Color')
plt.show()

# Show grayscal img
plt.imshow(black_n_white)
plt.title('Shahid Car Detector Gray Scale')
plt.show()
```

```python
# create our classifier

car_tracker = cv2.CascadeClassifier(classifier_file)
car_tracker
```

```python
# Detect cars

# multiscale means detect car of any size or of any scale
#  This function will return a rectangle with coordinates(x,y,w,h) around the detected cars .
cars = car_tracker.detectMultiScale(black_n_white)
cars
```

```python
car1 =cars[0]
car1
```

```python
# Unpacking array

(x, y, w, h) = car1
print("x", x)
print("y", y)
print("w", w)
print("h", h)
```

```python
# Draw rectangle around a car
# cv2.rectangle(image, start_point, end_point, color, thickness)
cv2.rectangle(
    black_n_white,
    (x, y),
    (x + w, y + h),
    (255,0,0), # BGR
    2
)
```

```python
plt.imshow(black_n_white)
plt.title('Shahid Car Detector Gray Scale detected')
plt.show()
```

```python
for(x, y, w, h) in cars:
    cv2.rectangle(
    black_n_white,
    (x, y),
    (x + w, y + h),
    (255,0,0), # BGR
    2
    )
```
